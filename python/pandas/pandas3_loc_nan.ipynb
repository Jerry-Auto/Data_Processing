这一节主要是对pandas进行索引和修改

```python
import pandas as pd
import numpy as np
dates=pd.date_range("20251001",periods=6)
df=pd.DataFrame(np.arange(24).reshape(6,4),index=dates,columns=['A','B','C','D'])
```

```python
df
```

```python
df.iloc[2,2]=111
df
```

```python
df.loc['20251004','B']=222
df
```

```python
df.loc[df.A > 9, 'B'] = 0
df
```

```python
df['F']=np.nan
df
```

```python
df['E']=pd.Series(np.arange(6),index=df.index)
df
```

对丢失掉的数据数据进行处理

```python
df.iloc[0,1]=np.nan
df.iloc[1,2]=np.nan
df
```

丢掉有nan的数据，可以是列，可以是行，自定义轴，且丢弃条件（how）有两种：有一个（any）或者全部(all)

```python
df=df.dropna(axis=1,how='all')
df
```

```python
print(df.dropna(axis=0,how='any'))
df
```

把nan填成给定值

```python
print(df.fillna(value=999))
df
```

```python
print(df.isnull())
print(np.any(df.isnull())==True)
```
