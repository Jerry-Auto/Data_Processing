这节介绍堆叠和数据合并

```python
import pandas as pd
import numpy as np
```

```python
df1=pd.DataFrame(np.ones((3,4))*0,columns=['a','b','c','d'])
df2=pd.DataFrame(np.ones((3,4))*1,columns=['a','b','c','d'])
df3=pd.DataFrame(np.ones((3,4))*2,columns=['a','b','c','d'])
print(df1)
print(df2)
print(df3)



```

在指定轴上进行堆叠，可选是否忽略原有的index标签，重新生成新的连续的index
concat只能对比轴的标签是否相同来进行堆叠，可以进行左右，上下堆叠

```python
res=pd.concat([df1,df2,df3],axis=0,ignore_index=True)
res
```

当标签不完全相同的两个pd进行堆叠或合并

```python
df1=pd.DataFrame(np.ones((3,4))*0,index=[1,2,3],columns=['a','b','c','d'])
df2=pd.DataFrame(np.ones((3,4))*1,index=[2,3,4],columns=['b','c','d','e'])
print(df1)
print(df2)
```

```python
res=pd.concat([df1,df2],axis=0,ignore_index=True)
res
```

```python
res=pd.concat([df1,df2],axis=0,ignore_index=True,join='inner')#默认是outer
res
```

左右堆叠时会出现的问题,只能丢掉没有的或者全都要，现在可以指定留下哪一个的index

```python
res=pd.concat([df1,df2],axis=1,ignore_index=True,join='outer')#默认是outer
res
```

```python
res=pd.concat([df1, df2],axis=1,ignore_index=True).reindex(df1.index)
res
```

另外一种合并方式merge  
只能对列，即左右进行堆叠  
特点是可以使用某一列的值是否相同作为堆叠条件，同时作为堆叠条件的那一列会进行合并  
合并的方式有5种  
left:左边的df的堆叠条件列的全部值+左右两边的df的堆叠条件列的交集  right同理  
inner outer一样的



```python
left = pd.DataFrame({
    "key1": ["K0", "K0", "K1", "K2"],
    "key2": ["K0", "K1", "K0", "K1"],
    "A": ["A0", "A1", "A2", "A3"],
    "B": ["B0", "B1", "B2", "B3"],
})


right = pd.DataFrame({
    "key1": ["K0", "K1", "K1", "K2"],
    "key2": ["K0", "K0", "K0", "K0"],
    "C": ["C0", "C1", "C2", "C3"],
    "D": ["D0", "D1", "D2", "D3"],
})


pd.merge(left, right, on=["key1", "key2"])#默认是inner取交集 00,10,10
```

```python
pd.merge(left, right, on=["key1", "key2"],how='left')#00,10,10+01,21
```

join  
使用的 key 可以和 concat() 一样，都是 index，也可以像 merge() 带一个 on="key" 去使用一个 column 作为索引  
默认基于索引（Index）进行左连接（how='left'）

```python
left = pd.DataFrame({
    "A": ["A0", "A1", "A2"], 
    "B": ["B0", "B1", "B2"]
}, index=["K0", "K1", "K2"])

print(left)

right = pd.DataFrame({
    "C": ["C0", "C2", "C3"], 
    "D": ["D0", "D2", "D3"]
}, index=["K0", "K2", "K3"])
print(right)

left.join(right)
```

```python
left.join(right,how='right')
```

堆叠条件是列时，使用on='key'（不推荐使用，推荐用merge）
on='key'仅用于左表列，右表必须为索引

```python
left = pd.DataFrame({
    "A": ["A0", "A1", "A2", "A3"],
    "B": ["B0", "B1", "B2", "B3"],
    "key": ["K0", "K1", "K0", "K1"],
})
print(left)
right = pd.DataFrame({
    "C": ["C0", "C1"],
    "D": ["D0", "D1"]
}, index=["K0", "K1"])
#堆叠条件是left的key列和right的index是否相同
print(right)
left.join(right, on="key")
```
