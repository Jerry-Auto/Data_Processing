# ILQGame 算法的详细推导与扩展

## 从泛函开始的推导（连续时间）

### 泛函的定义
在连续时间最优控制和博弈论中，智能体的成本函数可以视为一个泛函（functional），它是控制输入和状态轨迹的函数映射到实数。

对于单智能体最优控制，成本泛函为：
$$J[u] = \int_0^T L(t, x(t), u(t)) dt + \Phi(x(T))$$

其中 $L$ 是拉格朗日函数，$\Phi$ 是终端成本。

动态约束为：
$$\dot{x} = f(t, x(t), u(t)), \quad x(0) = x_0$$

目标是找到 $u(t)$ 使得 $J[u]$ 最小化。

### 变分原理
使用变分法，考虑变分 $\delta J = 0$。

引入拉格朗日乘子 $\lambda(t)$（伴随变量），定义哈密顿函数：
$$H(t, x, u, \lambda) = L(t, x, u) + \lambda^T f(t, x, u)$$

则欧拉-拉格朗日方程为：
$$\frac{\partial H}{\partial u} = 0, \quad \dot{\lambda} = -\frac{\partial H}{\partial x}, \quad \dot{x} = \frac{\partial H}{\partial \lambda}$$

终端条件：$\lambda(T) = \frac{\partial \Phi}{\partial x}(x(T))$。

**详细推导**：对于泛函 $J[u] = \int_0^T L(t, x, u) dt + \Phi(x(T))$，变分 $\delta J = \int_0^T \left( \frac{\partial L}{\partial x} \delta x + \frac{\partial L}{\partial u} \delta u \right) dt + \frac{\partial \Phi}{\partial x} \delta x(T)$。

使用分部积分和动态约束 $\delta \dot{x} = \frac{\partial f}{\partial x} \delta x + \frac{\partial f}{\partial u} \delta u$，引入 $\lambda$ 使得 $\delta J = 0$ 导致上述方程。

### 多智能体博弈的扩展
对于多智能体系统，每个智能体 $i$ 有自己的成本泛函：
$$J_i[u_1, \dots, u_N] = \int_0^T L_i(t, x, u_1, \dots, u_N) dt + \Phi_i(x(T))$$

共享动态：
$$\dot{x} = f(t, x, u_1, \dots, u_N)$$

纳什均衡定义为：策略组 $(u_1^*, \dots, u_N^*)$ 使得对于每个 $i$，$J_i(u_i^*, u_{-i}^*) \leq J_i(u_i, u_{-i}^*)$ 对于所有 $u_i$。

使用变分法，每个智能体最小化自己的 $J_i$，假设其他策略固定。

对于智能体 $i$，哈密顿函数为：
$$H_i(t, x, u_1, \dots, u_N, \lambda_i) = L_i(t, x, u_1, \dots, u_N) + \lambda_i^T f(t, x, u_1, \dots, u_N)$$

则：
$$\frac{\partial H_i}{\partial u_i} = 0, \quad \dot{\lambda_i} = -\frac{\partial H_i}{\partial x}, \quad \dot{x} = \frac{\partial H_i}{\partial \lambda_i}$$

但由于 $\dot{x}$ 共享，需要耦合伴随变量。

**详细推导**：对于每个 $i$，变分 $\delta J_i = \int_0^T \frac{\partial L_i}{\partial x} \delta x + \sum_j \frac{\partial L_i}{\partial u_j} \delta u_j dt + \frac{\partial \Phi_i}{\partial x} \delta x(T)$。

固定 $u_{-i}$，$\delta u_j = 0$ for $j \neq i$，则 $\delta J_i = \int_0^T \left( \frac{\partial L_i}{\partial x} \delta x + \frac{\partial L_i}{\partial u_i} \delta u_i \right) dt + \frac{\partial \Phi_i}{\partial x} \delta x(T)$。

使用分部积分和 $\delta \dot{x} = \frac{\partial f}{\partial x} \delta x + \frac{\partial f}{\partial u_i} \delta u_i$，引入 $\lambda_i$ 得到上述。

由于 $\dot{x}$ 由所有 $u_j$ 决定，$\lambda_i$ 耦合。

### 线性化和二次化
ILQGame 将非线性问题近似为线性二次。

假设当前策略 $\bar{u}_i(t)$，状态 $\bar{x}(t)$。

系统线性化：
$$f(t, x, u) \approx f(\bar{x}, \bar{u}) + \frac{\partial f}{\partial x} (\bar{x}, \bar{u}) (x - \bar{x}) + \sum_i \frac{\partial f}{\partial u_i} (\bar{x}, \bar{u}) (u_i - \bar{u}_i)$$

**详细推导**：泰勒展开 $f(x, u) = f(\bar{x}, \bar{u}) + \nabla_x f \cdot (x - \bar{x}) + \sum_i \nabla_{u_i} f \cdot (u_i - \bar{u}_i) + O(2)$。

忽略高阶项。

成本函数二次化：对 $L_i$ 进行二阶泰勒展开。

$$L_i(x, u) \approx L_i(\bar{x}, \bar{u}) + \nabla_{x} L_i \cdot (x - \bar{x}) + \sum_j \nabla_{u_j} L_i \cdot (u_j - \bar{u}_j) + \frac{1}{2} (x - \bar{x})^T \nabla_{xx}^2 L_i (x - \bar{x}) + \frac{1}{2} \sum_j \sum_k (u_j - \bar{u}_j)^T \nabla_{u_j u_k}^2 L_i (u_k - \bar{u}_k) + (x - \bar{x})^T \sum_j \nabla_{x u_j}^2 L_i (u_j - \bar{u}_j)$$

**详细推导**：二阶泰勒：$L(x, u) = L(\bar{x}, \bar{u}) + \nabla L \cdot \delta + \frac{1}{2} \delta^T H \delta$，其中 $\delta = (x - \bar{x}, u - \bar{u})$，H 是海森矩阵。

类似处理终端成本。

### 线性二次博弈的求解
近似后，问题变为：
$$\dot{x} = A(t) x + \sum_i B_i(t) u_i$$

$$J_i = \frac{1}{2} \int_0^T (x^T Q_i x + 2 x^T \sum_j S_{ij} u_j + \sum_j \sum_k u_j^T R_{ijk} u_k) dt + \frac{1}{2} x(T)^T P_i x(T)$$

使用动态规划，定义值函数 $V_i(t, x) = \frac{1}{2} x^T P_i(t) x$

哈密顿函数：
$$H_i = \frac{1}{2} x^T Q_i x + \sum_j \sum_k u_j^T R_{ijk} u_k + 2 x^T \sum_j S_{ij} u_j + P_i x \cdot (A x + \sum B_j u_j) + \dot{P_i} \frac{1}{2} x^T x$$

假设反馈策略 $u_j = K_j x$，代入得到 Riccati 方程。

### Riccati 方程的详细推导
Riccati 方程是从 HJB 方程推导出来的，用于求解值函数矩阵 $P_i(t)$。

HJB 方程：
$$\frac{\partial V_i}{\partial t} + \min_{u_i} \max_{u_{-i}} H_i = 0$$

假设 $V_i(t, x) = \frac{1}{2} x^T P_i(t) x$，则：
- $\frac{\partial V_i}{\partial t} = \frac{1}{2} x^T \dot{P_i} x$
- $\frac{\partial V_i}{\partial x} = P_i x$
$$
H_i = L_i + \frac{\partial V_i}{\partial x}^T \dot{x} + \frac{\partial V_i}{\partial t} = L_i + P_i x \cdot \dot{x} + \frac{1}{2} x^T \dot{P_i} x
$$
将 $L_i$ 和 $\dot{x}$ 代入：
$$
\begin{align}
L_i &= \frac{1}{2} x^T Q_i x + \sum_j \sum_k u_j^T R_{ijk} u_k + 2 x^T \sum_j S_{ij} u_j \\
\dot{x} &= A x + \sum_j B_j u_j
\end{align}
$$
所以
$$
\begin{align}
H_i &= \frac{1}{2} x^T Q_i x + \sum_j \sum_k u_j^T R_{ijk} u_k + 2 x^T \sum_j S_{ij} u_j + P_i x \cdot (A x + \sum_j B_j u_j) + \frac{1}{2} x^T \dot{P_i} x \\
&= \frac{1}{2} x^T Q_i x + \sum_j \sum_k u_j^T R_{ijk} u_k + 2 x^T \sum_j S_{ij} u_j + x^T P_i A x + x^T P_i \sum_j B_j u_j + \frac{1}{2} x^T \dot{P_i} x
\end{align}
$$

现在，$\min_{u_i} \max_{u_{-i}} H_i$ 意味着对于反馈策略 $u_j = K_j x$，H_i 成为 x 的二次函数，我们令其系数为零。

将 $u_j = K_j x$ 代入：
$$
\begin{align}
\sum_j \sum_k (K_j x)^T R_{ijk} (K_k x) &= \sum_j \sum_k x^T K_j^T R_{ijk} K_k x \\
2 x^T \sum_j S_{ij} (K_j x) &= 2 x^T \sum_j S_{ij} K_j x \\
x^T P_i \sum_j B_j (K_j x) &= x^T P_i \sum_j B_j K_j x
\end{align}
$$

所以
$$
\begin{align}
H_i &= \frac{1}{2} x^T \left( Q_i + 2 \sum_j \sum_k K_j^T R_{ijk} K_k + 2 P_i A + 2 P_i \sum_j B_j K_j + \dot{P_i} \right) x + x^T \left( 2 \sum_j S_{ij} K_j \right) x + \text{常数项}
\end{align}
$$

$H_i$ 是 $\frac{1}{2} x^T \Gamma x + \gamma^T x + \eta$，其中 $\Gamma$ 是二次项矩阵。

HJB 要求 $\Gamma = 0, \gamma = 0, \eta = 0$。

所以
$$
\begin{align}
\dot{P_i} + Q_i + 2 \sum_j \sum_k K_j^T R_{ijk} K_k + 2 P_i A + 2 P_i \sum_j B_j K_j &= 0 \\
\dot{P_i} &= - (Q_i + 2 \sum_j \sum_k K_j^T R_{ijk} K_k + 2 P_i A + 2 P_i \sum_j B_j K_j) 
\end{align}
$$

注意：$x^T P_i A x$ 的矩阵是 $P_i A$，但二次形式通常对称，所以标准形式为 $\dot{P_i} = - (A^T P_i + P_i A + \dots)$。

正确地，Riccati 方程为：
$$
\begin{align}
\dot{P_i} = - \left( A^T P_i + P_i A + \sum_j \sum_k K_j^T R_{ijk} K_k + Q_i + \sum_j (S_{ij}^T K_j + K_j^T S_{ij}) + P_i \sum_j B_j K_j + (\sum_j B_j K_j)^T P_i \right)
\end{align}
$$

终端条件：$P_i(T) = P_i$。

对于反馈增益，从 $\frac{\partial H_i}{\partial u_i} = 0$：
$$
\begin{align}
\frac{\partial H_i}{\partial u_i} &= 2 R_{i,ii} u_i + 2 \sum_{k \neq i} R_{i,ik} u_k + 2 S_{i,i}^T x + B_i^T P_i x = 0 \\
K_i &= - R_{i,ii}^{-1} \left( \sum_{k \neq i} R_{i,ik} K_k + S_{i,i}^T + \frac{1}{2} B_i^T P_i \right)
\end{align}
$$

### 迭代过程
1. 初始化 $\bar{u}_i^{(0)}$
2. 迭代：线性化、二次化、求解 LQ 博弈、更新策略，直到收敛。

**详细**：收敛条件 $\max_i \int_0^T \|u_i^{(m+1)} - u_i^{(m)}\|^2 dt < \epsilon$

## LQGame 的交互处理
在 LQGame 中，没有硬约束，交互通过成本函数的耦合项建模。

- **耦合项**：$R_{ijk}, S_{ij}$ 反映智能体对彼此的依赖。
- **碰撞避免**：添加距离惩罚 $(x_i - x_j)^T Q_{ij} (x_i - x_j)$。
- **跟车**：添加跟随项 $(v_i - v_j)^T R (v_i - v_j)$。
- **变道**：添加变道激励和冲突惩罚。

纳什均衡确保协调行为。

**详细**：例如，碰撞避免：$J_i$ 包含 $\frac{1}{2} ( \|x_i - x_j\|^2 - d_{\min}^2 )^2$，二次化后成为 $(x_i - x_j)^T Q (x_i - x_j)$。

## 离散时间 LQGame
动态：$x_{k+1} = A_k x_k + \sum B_{i,k} u_{i,k}$

成本：$J_i = \sum_{k=0}^{N-1} l_i + \phi_i$

值函数：$V_i(k, x) = \frac{1}{2} x^T P_{i,k} x$

Riccati：$P_{i,k} = Q_{i,k} + \sum_j \sum_m K_{j,k}^T R_{i,jm,k} K_{m,k} + \tilde{A}_k^T P_{i,k+1} \tilde{A}_k + \dots$

**详细推导**：Bellman $V_i(k) = \min_{u_i} \max_{u_{-i}} [l_i + V_i(k+1)]$

假设 $V_i(k+1) = \frac{1}{2} x_{k+1}^T P_{i,k+1} x_{k+1}$，$x_{k+1} = \tilde{A}_k x_k$

l_i 二次化，代入反馈，匹配。

反馈：$K_{i,k} = - R_{i,ii,k}^{-1} (\sum_{m \neq i} R_{i,im,k} K_{m,k} + S_{i,i,k}^T + \frac{1}{2} B_{i,k}^T P_{i,k+1} \tilde{A}_k)$

## 离散时间 ILQGame
扩展 ILQGame 到离散时间：

1. 初始化策略 $\bar{u}_{i,k}^{(0)}$
2. 迭代：
   - 模拟轨迹 $\bar{x}_k, \bar{u}_{i,k}$
   - 线性化：$x_{k+1} \approx A_k x_k + \sum B_{i,k} u_{i,k} + c_k$
   - 二次化成本：提取 $Q, R, S, P$
   - 求解离散 LQ 博弈得到 $K_{i,k}^{(m+1)}$
   - 更新 $\bar{u}_{i,k}^{(m+1)} = K_{i,k}^{(m+1)} \bar{x}_k$
3. 直到收敛。

适用于数字控制和仿真。
